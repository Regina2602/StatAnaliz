```python
import numpy as np
from scipy.stats import t
from scipy.stats import norm
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
import io
import requests
import seaborn as sns
import statistics
from scipy.stats import pearsonr
from sklearn.metrics import mean_squared_error, mean_absolute_error
import random
%matplotlib inline
```

**Пример 1**

```python
data = pd.read_csv('example.csv')
data.head()
```

```python
fonds = data['fonds']
product = data['product'] 
#plt.scatter(freedom, positive)
plt.scatter(fonds, product)
plt.xlabel('Стоимость основных фондов')
plt.ylabel('Средняя выработка на 1 работника')
```

```python
fonds = np.array(fonds).reshape(-1,1)
product = np.array(product).reshape(-1,1)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    fonds,
    product,
    test_size = 0.2
)
```

```python
model.fit(X_train, y_train)
```

```python
predicted_y = model.predict(X_test)
print(mean_squared_error(predicted_y, y_test))
print(mean_absolute_error(predicted_y, y_test))
```

```python
plt.xlabel('Стоимость основных фондов')
plt.ylabel('Средняя выработка на 1 работника')
plt.title('Линейная регрессия от одной переменной')
x = np.linspace(-50, 200, 100)
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(0,30)
plt.ylim(0,60)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

С увеличением стоимости основных фондов, увеличивается средняя выработка на одного работника.
